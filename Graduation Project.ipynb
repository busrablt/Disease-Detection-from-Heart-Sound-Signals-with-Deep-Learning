# Feature Extraction

```python
# All paths of mfcc data.
paths = ['AS_New_308/*.wav', 'MS_New_308/*.wav', 'MR_New_308/*.wav', 'MVP_New_308/*.wav', 'N_New_308/*.wav']
data=[]
entireMfcc = () # Tuple for concatenation proccess.
for path_i in range(len(paths)):
    data = glob.glob(paths[path_i]) # Get index of the path and set the data. 
    temp_mfcc=[]
    for i in range(len(data)):
        sample_rate, signal = scipy.io.wavfile.read(data[i])
        signal = np.float32(signal)
        mfccs = librosa.feature.mfcc(y=signal, sr=sample_rate, n_mfcc=40)
        mfccs = np.mean(mfccs.T,axis=0)
        temp_mfcc.append(np.asarray(mfccs).flatten())
    temp_mfcc = np.asarray(temp_mfcc) # temporary variable for adding the data to "entireMfcc"
    entireMfcc = entireMfcc + (temp_mfcc, ) # In each iteration, the data adding on cumulatively
```

```python
import pywt
```

```python
# All paths of dwt data.
paths = ['AS_New_308/*.wav', 'MS_New_308/*.wav', 'MR_New_308/*.wav', 'MVP_New_308/*.wav', 'N_New_308/*.wav']
data=[]
entireDwt = () # Tuple for concatenation proccess.
for path_i in range(len(paths)):
    data = glob.glob(paths[path_i]) # Get index of the path and set the data. 
    temp_dwt=[]
    for i in range(len(data)):
        sample_rate,signal=scipy.io.wavfile.read(data[i]) #Reading Audio
        signal=np.float32(signal)
        cA, cD=pywt.dwt(signal,'haar') #DWT cA->Approximation coeffs cD->Detailed coeffs
        dwts=pywt.idwt(cA, cD,'haar') #IDWT
        dwts = np.mean(dwts.T,axis=0)
        temp_dwt.append(np.asarray(dwts).flatten()) 
    temp_dwt = np.asarray(temp_dwt)
    entireDwt = entireDwt + (temp_dwt, )
```

```python
# Evaluating the ANN
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential # initialize neural network library
from keras.layers import Dense # build our layers library
def build_classifier():
    classifier = Sequential() # initialize neural network
    classifier.add(Dense(units = 256, kernel_initializer = 'uniform', activation = 'relu', input_dim = n_features))
    classifier.add(Dense(units = 128, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier, epochs = 100, batch_size=100)
accuracies = cross_val_score(estimator = classifier, X = x_test, y = y_test)
mean = accuracies.mean()
variance = accuracies.std()
print("Accuracy mean: "+ str(mean))
print("Accuracy variance: "+ str(variance))
```
